# Added value for Staff

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

m_cast = pd.read_csv('./data/Movie Cast.csv')
m_data = pd.read_csv('./data/Movie Data.csv')

print(m_cast.shape)
m_cast.head()
```

```python
m_cast.Type.value_counts()
```

```python
target_cols = [
    'Domestic Box Office Revenue (thousands of $)', 
    'International Box Office Revenue (thousands of $)'
]
budget = 'Budget (thousands of $)'

clean_target_cols = m_data[target_cols].fillna(0)

y = clean_target_cols.iloc[:,0] + clean_target_cols.iloc[:,1] - m_data[budget].fillna(0)
```

## Actors

```python
actor_movies = m_cast.loc[m_cast.Type == 'Actors'].groupby('Name').Movie.unique()
actor_movies = actor_movies.loc[actor_movies.map(len) >= 10]

def movies_to_mean_revenue(movies):
    return y.loc[m_data.Movie.isin(movies)].mean()

actors_revenue = actor_movies.map(movies_to_mean_revenue).to_frame()
actors_revenue.rename(columns={'Movie': 'Profit'}, inplace=True)
actors_revenue.sort_values('Profit', ascending=False)
```

## Staff

```python
staff_movies = m_cast.loc[m_cast.Type == 'Production crew'].groupby('Name').Movie.unique()
staff_movies = staff_movies.loc[staff_movies.map(len) >= 10]

def movies_to_mean_revenue(movies):
    return y.loc[m_data.Movie.isin(movies)].mean()

staff_revenue = staff_movies.map(movies_to_mean_revenue).to_frame()
staff_revenue.rename(columns={'Movie': 'Profit'}, inplace=True)
staff_revenue.sort_values('Profit', ascending=False)
```

## Without franchises

```python
actor_movies = m_cast.loc[m_cast.Type == 'Actors'].groupby('Name').Movie.unique()
actor_movies = actor_movies.loc[actor_movies.map(len) >= 10]

def movies_to_mean_revenue(movies):
    return y.loc[m_data.Movie.isin(movies) & m_data.Franchise.isnull()].mean()

actors_revenue = actor_movies.map(movies_to_mean_revenue).to_frame()
actors_revenue.rename(columns={'Movie': 'Profit'}, inplace=True)
actors_revenue.sort_values('Profit', ascending=False)
```

```python
actor_movies = m_cast.loc[m_cast.Type == 'Production crew'].groupby('Name').Movie.unique()
actor_movies = actor_movies.loc[actor_movies.map(len) >= 10]

def movies_to_mean_revenue(movies):
    return y.loc[m_data.Movie.isin(movies) & m_data.Franchise.isnull()].mean()

actors_revenue = actor_movies.map(movies_to_mean_revenue).to_frame()
actors_revenue.rename(columns={'Movie': 'Profit'}, inplace=True)
actors_revenue.sort_values('Profit', ascending=False)
```
